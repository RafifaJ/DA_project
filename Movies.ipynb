```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv("/content/movies.csv")
```

```python
df.head()
```

```python
df.tail()
```

###Clean data

```python
for col in df.columns:
 pct_missing = np.mean(df[col].isnull())
 print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(df['budget'].median())
df['gross'] = df['gross'].fillna(df['gross'].median())
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}% '.format(col, round(pct_missing*100)))
```

```python
df.info()
```

```python
df.describe()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype('Int64')
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
df.boxplot(column= ['gross'], vert = False)
```

```python
sns.regplot(x='gross', y='budget', data=df, scatter_kws={"color": "red"}, line_kws={"color": "purple"})
```

```python
sns.regplot(x='score', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "purple"})
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

plt.figure(figsize=(10,12)) # Set the figure size
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
